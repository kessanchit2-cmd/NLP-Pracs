```python
#imports
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import defaultdict
```

```python
#Download required datasets
# Download required datasets
nltk.download('punkt')
nltk.download('stopwords')
```

```python
#Input Text
text = """
Artificial Intelligence is transforming the world rapidly.
It is used in healthcare, education, finance, and transportation.
Machine learning is a subset of AI that enables systems to learn automatically.
AI improves efficiency and reduces human effort.
Many companies are investing heavily in AI technologies.
"""
```

```python
nltk.download('punkt_tab')
#tokenize sentences
sentences = sent_tokenize(text)
```

```python
# Tokenize sentences
sentences = sent_tokenize(text)
```

```python
# Tokenize words
words = word_tokenize(text.lower())


```

```python
# Remove stopwords
stop_words = set(stopwords.words("english"))
```

```python
# Word frequency
freq_table = defaultdict(int)
```

```python
for word in words:
    if word.isalnum() and word not in stop_words:
        freq_table[word] += 1
```

```python
# Score sentences
sentence_scores = defaultdict(int)
for sentence in sentences:
    for word, freq in freq_table.items():
        if word in sentence.lower():
            sentence_scores[sentence] += freq
```

```python
# Generate summary
summary_sentences = sorted(
    sentence_scores,
    key=sentence_scores.get,
    reverse=True
)[:2]

summary = " ".join(summary_sentences)
```

```python
# Output
print("Original Text:\n")
print(text)

print("\nSummary:\n")
print(summary)
```
