```python
#imports
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
```

```python
# Download tokenizer
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Input sentence
text = "Artificial Intelligence is changing the world with advanced technology"
```

```python
# Tokenize words
tokens = word_tokenize(text.lower())
```

```python
# Prepare data for Word2Vec
data = [tokens]
```

```python
# Train Word2Vec model
model = Word2Vec(
    sentences=data,
    vector_size=5,
    window=2,
    min_count=1,
    workers=1
)
```

```python
# Display word embeddings
print("Word Embeddings:\n")

for word in tokens:
    print(f"{word} : {model.wv[word]}")
```
