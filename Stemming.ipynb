```python
#import
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
# Download tokenizer
nltk.download('punkt_tab')
```

```python
# Create Porter Stemmer object
ps = PorterStemmer()
```

```python
# Input sentence
text = "playing played studies studying happiness running"
```

```python
# Tokenize words
words = word_tokenize(text)
print("Original Words and Stemmed Words:\n")
```

```python
# Perform stemming
for word in words:
    stem_word = ps.stem(word)
    print(f"{word} --> {stem_word}")
```
