```python
#import
import spacy
#load English NLP Model
nlp = spacy.load("en_core_web_sm")
```

```python

#Input text
Text="""Sundar Pichai is the CEO of Google.
He lives in California and visited India on 15 August 2025
"""
```

```python
#Process text
doc=nlp(Text)
```

```python
print("Named Entities:\n")

# Extract entities
for ent in doc.ents:
    print(f"Entity : {ent.text}")
    print(f"Label  : {ent.label_}")
    print(f"Start  : {ent.start_char}")
    print(f"End    : {ent.end_char}")
    print("-" * 30)
```
